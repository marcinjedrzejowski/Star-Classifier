# Dataset Description for Star Classification

The dataset under examination presents a detailed set of attributes aimed at characterizing celestial objects, specifically stars. Each row in the dataset corresponds to a unique object, with the following key attributes:

## Object Identification

- **Object Identifier (`obj_ID`):** A unique identifier referencing each object in the image catalog used by the Sloan Digital Sky Survey's (SDSS) Catalog Archive Server (CAS).

## Celestial Coordinates

- **Coordinates (`alpha` and `delta`):** Right Ascension and Declination angles at the J2000 epoch, providing precise celestial location information for each object.

## Photometric System Filters

- **Magnitudes (`u`, `g`, `r`, `i`, `z`):** Measurements in various filters spanning ultraviolet, green, red, near-infrared, and infrared wavelengths, respectively.

## Observation Details

- **Run Number (`run_ID`):** Identifies the specific scan or observational run during which the object was studied.
- **Rerun Number (`rerun_ID`):** Specifies the processing methodology applied to the image data, allowing for traceability in the analysis.

## Camera and Field Information

- **Camera Column (`cam_col`):** Identifies the scanline within a particular run, corresponding to the camera used.
- **Field Number (`field_ID`):** Provides a field-specific identifier, aiding in the categorization of observed regions.

## Spectroscopic Attributes

- **Spectroscopic Object ID (`spec_obj_ID`):** A unique identifier for optical spectroscopic objects. Notably, objects with the same `spec_obj_ID` across different observations share the same output class.

## Object Class

- **Object Class (`class`):** Indicates the overarching class of the object, distinguishing between galaxies, stars, or quasars.

## Redshift

- **Redshift (`redshift`):** A measure of the increase in wavelength, providing insights into the object's distance and motion.

## Spectrograph Details

- **Plate ID (`plate`):** Identifies each plate utilized in the SDSS, contributing to the spectroscopic analysis.
- **Modified Julian Date (`MJD`):** Indicates the date when a specific piece of SDSS data was acquired.
- **Fiber ID (`fiber_ID`):** Specifies the fiber responsible for directing light to the focal plane during each observation.

This dataset serves as a valuable resource for star classification, encompassing a wide array of attributes that collectively contribute to a comprehensive understanding of celestial objects. The `class` column, in particular, emerges as a pivotal feature for discerning the object type, while `redshift` provides essential information regarding the object's cosmological characteristics. The spectroscopic attributes further enhance the dataset's utility, facilitating in-depth analyses of stellar properties.

# Selected Columns for Analysis

After careful examination of the dataset, we have decided to focus our analysis on the following columns:

- **Ultraviolet Filter (`u`):** Represents the magnitude of the star in the ultraviolet filter of the photometric system.

- **Green Filter (`g`):** Signifies the magnitude of the star in the green filter of the photometric system.

- **Red Filter (`r`):** Denotes the magnitude of the star in the red filter of the photometric system.

- **Near Infrared Filter (`i`):** Represents the magnitude of the star in the near-infrared filter of the photometric system.

- **Infrared Filter (`z`):** Indicates the magnitude of the star in the infrared filter of the photometric system.

These selected columns provide valuable information about the brightness of stars across different wavelengths, allowing us to focus our analysis on specific photometric characteristics. This focused approach will streamline our investigation and enhance the clarity of our findings.


# Dataset Imbalance Analysis

Following a thorough analysis of the dataset, a notable observation has surfaced regarding its class distribution. It has been identified that the dataset is unbalanced, with a significant disparity in the number of galaxies compared to stars and quasars.

## Class Distribution

- **Galaxies:** The dataset contains a substantial number of entries corresponding to galaxies.

- **Stars:** The representation of stars in the dataset is comparatively lower.

- **Quasars:** Similar to stars, quasars exhibit a lower frequency in the dataset.

## Implications

The observed class imbalance could impact the robustness and generalization capability of our analyses and models. Specifically, the overrepresentation of galaxies may lead to a model bias towards galaxy-related patterns, potentially overshadowing the characteristics of stars and quasars.
